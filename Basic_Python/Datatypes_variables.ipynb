VARIABLES
-we have few rules to write a variable they are:

1.  variables should not start with a number if you want to start use underscore(_)


2. variable should be a single character or a single word but not multiple words.If you want to use multiple use  (_)
3. Python variables are totally case sensitive.


4.   Dont use punctuations as a variable like &,*,#,(,)....











```python
9=sainath #error because var should not start with number
print(9)
```

```python
_9='sainath'
print(_9)
```

# Type conversions:
we can convert one datatype into the another datatype in python.



```python
#from int to float datatype.
a=25
print(a)
print(type(a))
b=float(a)
print(b)
print(type(b))
```

```python
#from int to string datatype
a=25
print(a)
print(type(a))
b=str(a)
print(b)
print(type(b))
```

```python
#from float to int
a=25.25
print(a)
print(type(a))
b=int(a)
print(b)
print(type(b))
```

```python
#from float to string datatype
a=25.125
print(a)
print(type(a))
b=str(a)
print(b)
print(type(b))
```

```python
#from string to int datatype
a='4568'
print(a)
print(type(a))
b=int(a)
print(b)
print(type(b))
```

```python
#from string to float datatype
a='4568'
print(a)
print(type(a))
b=float(a)
print(b)
print(type(b))
```

```python
#from string to float datatype
a='sainath4568'
print(a)
print(type(a))
b=float(a) # here  we got error because we cannot change the character into a int or float..
print(b)
print(type(b))
```

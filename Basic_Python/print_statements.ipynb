```python
a=27
print('my age is:'+str(a))
```

```python
#In this we can use only string datatype to add more variables
a=27
b=50.25
c='sainath'
print('my name is '+str(a)+' my salery is:'+str(b)+' rupees my name is '+c)
```

TYPE:3

Using Identifiers:

    %d for int variable

    %f for float variable

     %s for string variable

```python
a=12
b=25.5
c='sainath'
print('my age is %d'%a)
print('my salary is %f'%b)
print('my name is %s'%c)
```

```python
#we can also write them in asingle line
a=12
b=25.5
c='sainath'
print('my age is %d my salary is %f my name is %s'%(a,b,c))
```
